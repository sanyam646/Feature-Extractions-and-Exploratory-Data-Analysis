```python
#Now, Starting with comparing Pairwise Mean ZRC values, We will do so by using Euclidean distances between means of ZCR values
# We will use heatmap to analyse the same.
```

```python
from sklearn.metrics.pairwise import euclidean_distances

mean_zcr_values = []

for file in audio_files:
    audio, sr = librosa.load(file)
    zcr = librosa.feature.zero_crossing_rate(audio)
    mean_zcr = np.mean(zcr)
    mean_zcr_values.append(mean_zcr)

# Calculate pairwise distances between mean ZCR values
distance_matrix = euclidean_distances(np.array(mean_zcr_values).reshape(-1, 1))

# Visualize the distance matrix as a heatmap
sns.heatmap(distance_matrix, annot=True, xticklabels=audio_files, yticklabels=audio_files)
plt.title('Pairwise Distance Matrix of Mean ZCR Values')
plt.show()
```

```python
#From the above heatmap of the distances, we can incur following:
#Number 5 audio have similar ZCR mean values among each other and so do Number 9 audio.
#If we will compare Number 9 and Number 5 audio files data, they have greater differences in their value or we can say greater mean distances between the two.
```

```python
#Now we will use mean values of RMS(energy) to see if we can get any information from that.
```

```python
mean_rms_values = []

for file in audio_files:
    audio, sr = librosa.load(file)
    rms = librosa.feature.rms(y=audio)[0]
    mean_rms = np.mean(rms)
    mean_rms_values.append(mean_rms)

# Calculate pairwise distances between mean RMS values
distance_matrix_rms = euclidean_distances(np.array(mean_rms_values).reshape(-1, 1))

# Visualize the distance matrix as a heatmap
sns.heatmap(distance_matrix_rms, annot=True, xticklabels=audio_files, yticklabels=audio_files)
plt.title('Pairwise Distance Matrix of Mean RMS Values')
plt.show()
```

```python
#We can see there is no much difference or distance between mean RMS values of Nine and five. So, it is not
#NOT that great method for Feature Extraction.
```

```python
#FINALLY HISTOGRAM ANALYSIS OF ALL THE FEATURES to check what suits us the best.
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

def extract_features(file_path):

    y, sr = librosa.load(file_path)


    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    zero_crossings = librosa.feature.zero_crossing_rate(y)
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
    rms_energy = librosa.feature.rms(y=y)[0]
    spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)[0]
    pitches, magnitudes = librosa.core.pitch.piptrack(y=y, sr=sr)


    pitch = np.mean(pitches[pitches > 0])


    spectral_flux = librosa.onset.onset_strength(y=y, sr=sr)

    return mfccs, zero_crossings, spectral_centroid, rms_energy, spectral_rolloff, pitch, spectral_flux


file_paths = ["/content/5_yweweler_2.wav", "/content/5_yweweler_11.wav", "/content/5_yweweler_10.wav",
              "/content/9_yweweler_4 (1).wav", "/content/9_yweweler_5.wav", "/content/9_yweweler_6.wav"]



features_list = [extract_features(file_path) for file_path in file_paths]

def descriptive_stats(feature_index):
    feature_values_five = [feature_set[feature_index].mean() for feature_set in features_list[:3]]
    feature_values_nine = [feature_set[feature_index].mean() for feature_set in features_list[3:]]

    mean_five = np.mean(feature_values_five)
    std_five = np.std(feature_values_five)
    mean_nine = np.mean(feature_values_nine)
    std_nine = np.std(feature_values_nine)

    return feature_values_five, feature_values_nine, mean_five, std_five, mean_nine, std_nine


feature_names = ["MFCCs", "Zero Crossing Rate", "Spectral Centroid", "RMS Energy", "Spectral Rolloff", "Pitch", "Spectral Flux"]

for i, feature_name in enumerate(feature_names):
    feature_values_five, feature_values_nine, mean_five, std_five, mean_nine, std_nine = descriptive_stats(i)

    print(f"Comparison for {feature_name}:")
    print(f"Mean of 'five' category: {mean_five}, Standard Deviation: {std_five}")
    print(f"Mean of 'nine' category: {mean_nine}, Standard Deviation: {std_nine}")


    plt.hist(feature_values_five, bins=20, alpha=0.5, label='five')
    plt.hist(feature_values_nine, bins=20, alpha=0.5, label='nine')
    plt.title(f'Histogram Comparison for {feature_name}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
#zero crossing line is giving distinct lines for all 6 audio file data with histogram bars of fives closer and nine closer being far apart from each otehr
#We prefer only those features for classification which have bars of five and nine away from each other i.e. different set of values.
#Pitch is also a good feature for classification as it has bars of nines and fives far away.
```
